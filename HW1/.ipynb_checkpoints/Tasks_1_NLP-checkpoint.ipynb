Завдання 1. Розширити корпус стоп слів стоп словами української мови.

```python
import nltk
import os
import re
import math
```

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def tf_calc(occur,all_words):
    """Calculation of word occurence proportion"""
    return occur/all_words

def idf_calc(occur,all_docs):
    """Calculation of docs with certain word proportion"""
    return math.log(all_docs/occur)

def tfidf_calc(tf, idf):
    """Calculation TFIDF of word's occurence"""
    return tf * idf

#CREATED TO ORDER THE NAMES BY UKRAINIAN ALPHABET
UKR_LANGUAGE_ORDER = ['а','б','в','г','ґ','д','е','є','ж','з','и','і','i','ї','й','к','л',
                      'м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ь','ю','я']
```

```python
"""
SOURCE: a) https://github.com/brown-uk/corpus
"""    

MAIN_PATH = 'corpus-master/data'

STOPWORDS_PATH = '/home/ostapkharysh/nltk_data/corpora/stopwords/'

DOCUMENTS_COUNT = 0

DATA_DIRECTORIES = ['/good', '/so-so', '/bad'] 

THRESHHOLD = 0.001

word_tf = dict()


# CALCULATION OF THE WORD OCCURENCE FOR TFIDF ALGORITHM
for directory in DATA_DIRECTORIES:
    for text_file in os.listdir(MAIN_PATH + directory):
        file_words = dict()
        DOCUMENTS_COUNT+=1
        with open(MAIN_PATH+directory+'/'+text_file, 'r',encoding='utf-8', errors='ignore') as text_data: #codecs.
            for line in text_data:
                line = remove_html_tags(line)
                text = line.strip().split(" ")
                for el in text:
                    el = re.sub(r'!+(?=.*\!")','',el)
                    el = el.replace(',', '')
                    el = el.replace('.', '')
                    el = el.replace('«', '')
                    el = el.replace('»', '')
                    el = el.lower()
                    if el:
                        if el in file_words:
                            file_words[el] += 1
                        else:
                            file_words[el] = 1
        #CALCULATING TF                   
        for word in list(file_words.keys()):
            if word in word_tf.keys():
                word_tf[word].append(tf_calc(file_words[word], sum(file_words.values())))
            else: 
                word_tf[word] = [tf_calc(file_words[word], sum(file_words.values()))]
```

```python
"ALL WORDS: " + str(len(word_tf.keys()))
```

```python
word_tf
```

```python
word_idf = dict()

# CALCULATING IDF
for word in word_tf.keys():
    word_idf[word] = idf_calc(len(word_tf[word]), DOCUMENTS_COUNT)
```

```python
word_idf
```

```python
DOCUMENTS_COUNT # NUMBER OF DOCUMENTS USED FOR TFIDF ALORITHM
```

```python
word_tfidf = dict()

# CALCULATING TFIDF
for key in word_tf.keys():
    word_tfidf[key] =  sum([el*word_idf[key] for el in word_tf[key]])/DOCUMENTS_COUNT
```

```python
word_tfidf
```

```python
#STOPWORDS WITH POINTS HIGHER THAN THRESHHOLD ASSIGNED ARE ADDED TO THE NLTK STOPWORDS
with open(STOPWORDS_PATH+'ukrainian', 'w') as ukr_stopwords:
    for el in word_tfidf.keys():
        if word_tfidf[el] > THRESHHOLD:
            ukr_stopwords.write(el +'\n')
```

Завдання 2. Розширити корпус імен іменами української мови. Обов'язково вказати джерело даних з яких формується списки імен.

```python
import urllib.request, urllib.parse, urllib.error, requests 
import sys
from bs4 import BeautifulSoup as BS

"""SOURCE: http://database.ukrcensus.gov.ua/dw_name/main.asp"""

BASE_URL = "http://database.ukrcensus.gov.ua/dw_name/tlum.asp?nom_zag="

UKR_NAMES = dict()
```

```python
UKR_NAME_TYPES = ["slav", "greek", "latin", "eastern", "jewish", "european"]
GENDER = ['ukr_male', 'ukr_female']
UKR_NAMES = dict()


for el in GENDER:
    UKR_NAMES[el] = []

#SORTING UKRAINIAN MALE AND FEMALE NAMES
for i in range(len(UKR_NAME_TYPES)):
    switch = 0
    previous =''
    r = requests.get(BASE_URL+str(i+1))
    soup = BS(r.content, 'html.parser')
    span_list = soup.find_all('span', {'class':"textOR"})
    #custom switcher,based on website specifics: no tag difference between male and female. 
    for el in span_list:
        name = el.text.split(" ")[0].split()[0]
        if previous and switch == 0:
            if UKR_LANGUAGE_ORDER.index(list(previous)[0].lower())> UKR_LANGUAGE_ORDER.index(list(name)[0].lower()):                                                                                                   
                switch += 1
        previous = name
        UKR_NAMES[GENDER[switch]].append(name)
```

```python
NAMES_PATH = '/home/ostapkharysh/nltk_data/corpora/names/'

#ALPHABETICAL SORTING OF NAMES AND FILE RECORDING
for el in GENDER:
    with open(el +'.txt', 'w') as ukr_names_f:
        for name in sorted(list(set(UKR_NAMES[el])), reverse=False):
            ukr_names_f.write(NAMES_PATH+name +'\n')
```

```python
class ukr_nltk:
    #NLTK WRAPPER TO WORK WITH UKRAINIAN MALE-FEMALE NAMES
    
    """
    Creating the corpus of ukrainian female and male names
    """
    
    def __fix_apostrophe_bag(self, strings_list):
        """
        Fixing apostrophe and dash name splitting of nltk.corpus.reader
        """
        correct_list = list()
        for i in range(len(strings_list)):
            if len(strings_list)-1 > i and strings_list[i+1] in ["’", "'", '-']: 
                    correct_list.append(strings_list[i]+strings_list[i+1]+strings_list[i+2])
            elif strings_list[i] in ["’", "'", '-'] or strings_list[i-1] in ["’", "'", '-']:
                pass
            else: 
                correct_list.append(strings_list[i])
        return correct_list
    
    def __init__(self, filepath, file_ids, encoding='utf-8'):
        """
        Initializing values of male and female names
        """
        #self.ukr_male = self.__fetch_names('ukr_male.txt')   # if not using nltk.corpus.reader
        self.__cor_reader = nltk.corpus.reader.plaintext.PlaintextCorpusReader(filepath, file_ids) 
        self.ukr_male = self.__fix_apostrophe_bag(self.__cor_reader.sents()[0])
        self.ukr_female = self.__fix_apostrophe_bag(self.__cor_reader.sents()[1])
```

```python
NAMES_PATH = '/home/ostapkharysh/nltk_data/corpora/names/'
files = ['ukr_male.txt', 'ukr_female.txt']
d = ukr_nltk(NAMES_PATH, files)

print("NUMBER OF NAMES:")
print('MALE: '+ str(len(d.ukr_male)))
print('FEMALE: '+ str(len(d.ukr_female)))
```
