Завдання 1. Пошук електронних адрес в тексті.

```python
import csv
import re
import regex
```

```python
def get_emails(data):
        """
        Emails are filtered from the noisy csv file
        """
        email_p2 = re.compile(r'[\w\.-]+@[\w\.-]+')
        reader = csv.reader(open(data, 'r'), delimiter= ",",quotechar='|')
        email = set()
        for line in list(reader):
            for el in line:
                el = re.sub(r'[А-Яа-я]', '', el)
                em = email_p2.findall(el)
                if em:
                    email.add(em[0])
        return email
    
def find_romaniuk_at_ucu(data):
    rom_data =list()
    rom = r'^[\w*]{0,1}\.*roman[i|y]*uk@ucu.edu.ua'
    for el in data:
        if re.findall(rom, el):
            rom_data.append(el)
    return rom_data, len(rom_data)
    
```

```python
data = list(get_emails('google.csv'))
rom_surnames, count = find_romaniuk_at_ucu(data)

print('Number of registered people with "romaniuk, romanuk" or "romanyuk" surnames: {0}\n\nYou could find them here: \n {1}'.format(count, rom_surnames))
```
